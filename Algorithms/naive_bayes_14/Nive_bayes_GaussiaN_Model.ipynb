```python
import pandas as pd
df = pd.read_csv("titanic (1).csv")
df.head()
```

```python
#df.drop(['Name','Pclass','Parch','Ticket','Embarked','PassengerId','Cabin'],axis='columns',inplace=True)
```

```python
df.drop(['Cabin'],axis='columns',inplace=True)
```

```python
df
```

```python
dummies = pd.get_dummies(df.Sex)
dummies.head(3)
```

```python
final = pd.concat([dummies,df],axis='columns')
final
```

```python
final.drop(['Sex'],axis='columns',inplace=True)
```

```python
final.drop(['SibSp'],axis='columns',inplace=True)
```

```python
final.drop(['female'],axis='columns',inplace=True)
```

```python
final
```

```python
final.isna().any()
```

```python
df_final = final.fillna(df['Age'].median())
```

```python
df_final.isna().any()
```

```python
inputs = df_final.drop(['Survived'],axis='columns')
target = df_final.Survived
#target
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
Nb = GaussianNB()
Nb.fit(X_train,y_train)
```

```python
Nb.score(X_test,y_test)
```
